```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_excel('WorldBank Data.xlsx')
data.head()
```

```python
data.columns
```

```python
data.describe()
```

```python
column_name = 'Life expectancy at birth (years)'

# Plot the Histogram
plt.figure(figsize=(10, 6))
plt.hist(data[column_name], bins=20, color='violet', edgecolor='black', alpha=0.7)
plt.xlabel(column_name)
plt.ylabel('Frequency')
plt.title('Distribution of Life Expectancy at Birth')
plt.grid(True)
plt.show()
```

```python
column_name = 'Region'

# Plot the Bar Chart
plt.figure(figsize=(10, 6))
data[column_name].value_counts().plot(kind='bar', color='skyblue')
plt.xlabel(column_name)
plt.ylabel('Count')
plt.title('Distribution of Regions')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Plot the Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(data['GDP per capita (USD)'], data['Life expectancy at birth (years)'], color='skyblue')
plt.xlabel('GDP per capita (USD)')
plt.ylabel('Life expectancy at birth (years)')
plt.title('Scatter Plot: GDP per capita vs Life Expectancy')
plt.grid(True)
plt.show()
```
